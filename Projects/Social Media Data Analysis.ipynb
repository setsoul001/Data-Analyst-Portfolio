# Social Media Data Analysis

```python
#importing necessary libraries

import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Generate random data for the social media dataabs

categories = ["Food", "Travel", "Fashion", "Fitness", "Music", "Culture", "Family", "Health"]

categories
```

```python
likes = [random.randint(50, 50000) for like in range(500)]

likes
```

```python
date = pd.date_range("2021-06-27", periods = 500)
date
```

```python
impressions = []
for i in range(500):
    avg = [random.randint(10,50) for i in range(4)]
    avg_sum = round(likes[i] * (sum(avg)/5))
    impressions.append(avg_sum)

impressions
```

```python
clicks = []
for i in range(500):
    avg = [random.randint(170, 400) for i in range(4)]
    avg_sum = round(impressions[i] / (sum(avg)/7))
    clicks.append(avg_sum)
clicks
```

```python
category = [random.choice(categories) for category in range(500)]
category
```

```python
sm_data = {"date": date, "category": category, "impressions": impressions, "likes": likes,  "clicks": clicks}
sm_data
```

```python
# Load the data into a Pandas DataFrame and Explore the data

sm_data = pd.DataFrame(sm_data)
sm_data
```

### EDA

```python
sm_data.info()
```

```python
sm_data.describe()
```

```python
# count of each cacategory

sm_data["category"].value_counts()
```

```python
# total impressions, likes and clicks by cateory

cat_sum = sm_data[["impressions","likes","clicks"]].groupby(category).sum()

cat_sum
```

```python
# like to impression ratio

LIR = (cat_sum["likes"]/cat_sum["impressions"])*100

LIR.sort_values()
```

```python
# Total Engagement Rate (by Impressions)

TER = ((cat_sum["likes"]+ cat_sum["clicks"])/cat_sum["impressions"]) * 100

TER.sort_values()
```

```python
#Average Engagement Rate per Category

cat_mean = sm_data[["impressions","likes","clicks"]].groupby(category).mean()

cat_mean.reset_index()
```

```python
# Click Through Rate

CTR = ( cat_sum["clicks"]/cat_sum["impressions"]) * 100

CTR.sort_values()
```

```python
# Action intent ratio

AIR = (cat_sum["clicks"]/cat_sum["likes"])

AIR.sort_values()
```

```python
# Visualize and Analyze the data

sns.histplot(data = sm_data, x = "category", hue = "category", legend = False)
```

```python
num_data = sm_data.select_dtypes(np.number)
num_cols = list(num_data.columns)
num_cols
```

```python
# 1. Create a figure and a set of subplots (1 row, multiple columns)
fig, axes = plt.subplots(1, len(num_cols), figsize=(16, 5))

for i, col in enumerate(num_cols):
    # 2. Prepare the data (reset_index is needed for Seaborn to see 'categories')
    average = sm_data[["category", col]].groupby("category").mean().reset_index()
    
    # 3. Plot specifically on the axis at index i
    sns.barplot(data=average, x="category", y=col, hue="category", ax=axes[i])
    
    # 4. Give each specific plot a title
    axes[i].set_title(f"Average {col}")

# 5. Clean up layout so labels don't overlap
plt.tight_layout()
plt.show()
```

```python
# 1. Melt the data so all numerical columns are in one 'Metric' column
df_melted = sm_data.melt(id_vars='category', value_vars=num_cols, var_name='Metric', value_name='Value')

# This replaces the loop and the manual FacetGrid setup
sns.catplot(
    data=df_melted, 
    x="category", 
    y="Value", 
    col="Metric", 
    kind="bar", 
    col_wrap=3,      # Wrap to new row after 3 plots
    sharey=False,
    hue = "category"
)
```

```python
# 1. Create a figure and a set of subplots (1 row, multiple columns)
fig, axes = plt.subplots(1, len(num_cols), figsize=(16, 5))

for i, col in enumerate(num_cols):
    # 3. Plot specifically on the axis at index i
    sns.boxplot(data=sm_data, x="category", y=col, hue="category", ax=axes[i])
    # 4. Give each specific plot a title
    axes[i].set_title(f"Average {col}")

# 5. Clean up layout so labels don't overlap
plt.tight_layout()
plt.show()

#sns.boxplot(data = sm_data, x = "categories", y = "likes", hue = "categories")
```
