```python
# Check for outliers

numeric_cols = ['Open Price', 'Close Price', 'High Price', 'Low Price',
                'Volume Traded', 'Market Cap', 'PE Ratio', 'Dividend Yield', 'EPS']

for name, df in [('June', june_stock), ('July', july_stock), ('Aug', aug_stock)]:
    print(f"\n === {name} outliers ===")
    for col in numeric_cols:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        outliers = df[(df[col] < Q1 - 1.5*IQR) | (df[col] > Q3 + 1.5*IQR)]
        if len(outliers) > 0:
            print(f" {col}: {len(outliers)} outliers")
        else:
            print(f" {col}: 0 outliers")
```

Varying range of outliers are present in most numeric columns in each dataset. The reason being each columns contain numeric values of various tickers with varying worth; eg. NVDIA stocks does not have the same price range as VZ stocks.

# Data Cleaning and Preparation

#### Conversion of date columns to datetime format

```python
june_stock["Date"] = pd.to_datetime(june_stock['Date'], dayfirst=True)

june_stock["Date"]
```

```python
july_stock["Date"] = pd.to_datetime(july_stock["Date"], format = "mixed")

july_stock["Date"]
```

```python
aug_stock["Date"] = pd.to_datetime(aug_stock["Date"], format = "mixed")

aug_stock["Date"]
```

```python
# Join all 3 tables together

stocks = pd.concat([july_stock, june_stock, aug_stock], axis=0)

stocks
```

#### Check for duplicates

The merged table has 8650 rows × 14 columns

```python
stocks[["Date", "Ticker", "Open Price"]].duplicated().sum()
```

```python
stocks[stocks[["Date", "Ticker", "Open Price"]].duplicated()]
```

```python
stocks[(stocks["Date"] == "2025-07-14") & (stocks["Ticker"] == "NKE")]
```

While checking 3 core variable: Date, Ticker and Open Price, only 1 duplicate was discovered. Upon further check it was discovered that only those 3 variables are duplicated, all other variable were different. 

```python
stocks[["Date", "Ticker"]].duplicated().sum()
```

A check for just duplicated Dates and Tickers was carried out. 1931 rows of data were duplicated with same Ticker recorded with different opening and closing price on the same day. These posed a problem that need to be solved.

```python
# Remove duplicates

stocks = stocks.sort_values(['Date', 'Ticker', 'Volume Traded'], ascending=[True, True, False])
stocks = stocks.drop_duplicates(subset=['Date', 'Ticker'], keep='first')
stocks
```

The cleaned dataset contain 6719 rows × 14 columns as opposed to 8650 rows × 14 columns from the direct merge of all 3 datasets.

```python
stocks[(stocks["Date"] == "2025-07-15") & (stocks["Ticker"] == "NKE")]
```

#### Features Engineering

```python
stocks["Daily returns(%)"] = (((stocks["Close Price"] - stocks["Open Price"]) / stocks["Open Price"]) * 100).round(2)

stocks["Daily returns(%)"]
```

```python
stocks["Price Range"] = stocks["High Price"] - stocks["Low Price"]

stocks["Price Range"]
```

```python
stocks["Price Range (%)"] = (stocks["Price Range"]/stocks["Open Price"]*100
).round(2)
stocks["Price Range (%)"]
```

```python
avg_vol =  stocks.groupby('Ticker')['Volume Traded'].transform("mean")
stocks['relative_volume'] = stocks['Volume Traded'] / avg_vol

stocks['relative_volume']
```

```python
stocks["Source Month"] = stocks["Date"].dt.month_name()

stocks["Source Month"]
```

# Exploratory Data Analysis

```python
# Plot closing price distributions by sector 

fig, axes = plt.subplots(1, 2, figsize=(18, 6))

# Left: boxplot of raw close prices
sns.boxplot(data=stocks, x='Sector', y='Close Price', ax=axes[0], hue = "Sector")
axes[0].set_title('Close price distribution by sector', fontweight = "bold")
axes[0].set_xlabel(' ')
axes[0].tick_params(axis='x', rotation=90)

# Right: boxplot of daily returns %
sns.boxplot(data=stocks, x='Sector', y='Daily returns(%)', ax=axes[1], hue = "Sector")
axes[1].set_title('Daily return (%) distribution by sector', fontweight = "bold")
axes[1].set_xlabel(' ')
axes[1].axhline(0, color='red', linewidth=0.8, linestyle='--')
axes[1].tick_params(axis='x', rotation=90)

plt.tight_layout()
plt.show()
```

The plots show the distribution of close price and daily return by sector.

The Close price distribution by sector reveal that the median close price of all the sectors lies between `$100` and `$250`, the most volatile sectors are Technology, Healthcare and Financials, and Consumer Discretionary has the highest and largest range of outliers with close prices ranging between `$3500` and `$4000`.

The daily return distribution by sector reveal that the median daily return percentage is about `0%`and the Energy and Consumer Discretionary sectors have the widest range of daily returns by percentage.

```python
# Plot trading volume trends over time

from matplotlib.ticker import MaxNLocator
fig, axes = plt.subplots(1,2, figsize = (18,6))

# Left: bar plot of trade volume per day
sns.barplot(data=stocks, y = "Volume Traded", x = "Date", ax = axes[0])
axes[0].set_title("\nTrading Volume over time\n", fontweight = "bold")
axes[0].set_xlabel(' ')
axes[0].tick_params(axis='x', rotation=90)
axes[0].xaxis.set_major_locator(MaxNLocator(nbins=10))

# Right: line plot showing trade volume over time
sns.lineplot(data = stocks, x = "Date", y = "Volume Traded", ax = axes[1])
axes[1].set_title("\nTrading Volume over time\n", fontweight = "bold")
axes[1].tick_params(axis = "x", rotation = 90)
axes[1].axhline(stocks["Volume Traded"].mean(), color='red', linewidth=0.8, linestyle='--')
axes[1].set_xlabel(" ")
axes[1].xaxis.set_major_locator(MaxNLocator(nbins=10))
plt.grid(True, linestyle=':', linewidth='0.5', color='gray')

plt.show()
```

Both plot tell one story. The total volume traded from June 1, 2025 to June 21, 2025 ranged between `$10,000,000` and `$16,000,000`, the total trading volume spiked and ranged from `$15,000,000` to `$22,000,000`.

```python
stocks.groupby("Sector")[["PE Ratio", "EPS", "Dividend Yield", "Close Price"]].corr()
```

```python
# Plot heatmap showing correlation between PE Ratio, EPS, Dividend Yield and Close Price in each sector

metrics = ["PE Ratio", "EPS", "Dividend Yield", "Close Price"]

sectors_corr = stocks.groupby("Sector")[metrics].corr()

sectors = stocks["Sector"].unique()

fig, axes = plt.subplots(2, 6, figsize = (22,10), sharey=True)
fig.subplots_adjust(hspace=0.5)

for name, sector in enumerate(sectors):
    if name < 6:
        mat = sectors_corr.loc[sector]
        sns.heatmap(mat, ax = axes[0, name], annot=True, cbar = False, cmap = "RdBu", vmin=-1, vmax=1)
        axes[0,name].set_title(sector, fontsize=12, fontweight='bold')
    elif name >= 6:
        names = name - (6)
        mat = sectors_corr.loc[sector]
        sns.heatmap(mat, ax = axes[1, names], annot=True, cbar = False, cmap = "RdBu", vmin=-1, vmax=1 )
        axes[1,names].set_title(sector, fontsize=12, fontweight='bold')
fig.delaxes(axes[1, 5])
```

All sector show a strong correlation between Earning Per Share (EPS) and Close Price. Technology, Healthcare, Communication sectors show a moderately strong correlation between PE ratio and close price, while Financials, Materials, Energy, Utilities show a weak correlation and Real Estate, Consumer Discretionary, Industrials show negative correlation between PE ratio and close price. All sector except the energy sector show a negative correlation between PE ratio and dividend yield.

# Summary

Between June and July 2025, Utilities delivered the strongest average daily returns across all sectors, while Communication Services was the clear laggard, followed closely by Consumer Staples and Healthcare — all three averaging negative daily returns over the period.

Materials and Energy were the most volatile sectors by intraday price range, with both averaging above 3.5% daily swings, followed by Technology. Despite its high volatility, Energy still posted positive average returns, suggesting that its price swings skewed upward over the period.

Among the 82 core tickers, INTU, PYPL, and NOW were the strongest performers by mean daily return, while ELV, COST, and PFE lagged — all three underperformers sit in defensive sectors that trailed the broader market during this period.
